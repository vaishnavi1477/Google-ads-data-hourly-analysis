```python
dataset=file.fillna(file.mean())
```

```python
# data preprocessing
```

```python
column_to_drop=['Sr no','CTR','CPC','Warm Leads']
```

```python
newfile=dataset.drop(column_to_drop, axis=1)
```

```python
newfile
```

```python
#discriptive analysis
```

```python
newfile.sum()
```

```python
# Calculating average value
```

```python
newfile.mean()
```

```python
X=newfile.iloc[:,:-3].values
Y=newfile.iloc[:,-3].values
```

```python
X
```

```python
Y
```

```python
#decision tree classifier
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.05, random_state=0)
```

```python
 print(X_train)
```

```python
print(Y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT= DecisionTreeClassifier(criterion="gini", random_state=0)
```

```python
DT.fit(X_train,Y_train)
```

```python
print(DT.predict(sc.transform([[652,26]])))
```

```python
# Predicting the Test set results
```

```python
y_pred = DT.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), Y_test.reshape(len(Y_test),1)),1))
```

```python
# Making the Confusion Matrix
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, y_pred)
print(cm)
accuracy_score(Y_test, y_pred)
```

```python
# Visualising the Training set results
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, DT.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Decision Tree Classification (Training set)')
plt.xlabel('Impressions')
plt.ylabel('Cost')
plt.legend()
plt.show()
```
